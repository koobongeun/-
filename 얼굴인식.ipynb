```python
import dlib
import cv2 as cv
import numpy as np

# 얼굴 검출을 위해 디폴트 얼굴 검출기를 사용
detector = dlib.get_frontal_face_detector()

#검출된 얼굴에서 눈,코,입 같은 랜드마크를 찾기 위해 사용할 학습모델을 로드
# 총 68개의 위치를 찾아주는 학습모델을 사용
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')

#웹캠으로부터 영상을 가져와 입력으로 사용
cap = cv.VideoCapture(0)

# 눈, 코, 입, 턱선에 대한 번호가 정해져 있음
ALL = list(range(0,68))
RIGHT_EYEBROW = list(range(17,22))
LEFT_EYEBROW = list(range(22,27))
RIGHT_EYE = list(range(36,42))
LEFT_EYE = list(range(42,48))
NOSE = list(range(27,36))
MOUTH_OUTLINE = list(range(48,61))
MOUTH_INNER = list(range(61,68))
JAWLINE = list(range(0,17))

#전체 랜드마크를 보여주는 초기값
index = ALL

#웹캠으로부터 입력을 받으려면 무한 반복
while True :
    # 웹캠으로부터 이미지를 가져와서 그레이스케일로 변환
    ret, img_frame = cap.read()
    
    img_gray = cv.cvtColor(img_frame, cv.COLOR_BGR2GRAY)
    
    #주어진 이미지에서 얼굴을 검출
    dets = detector(img_gray, 1)
    
    #업샘플링(이미지를 확대), 발견된 얼굴 수 만큼 반복
    for face in dets :
        
        #주어진 이미지 img_frame의 검출된 얼굴 영역 face에서 랜드마크를 검출, 68개 점 찾기
        shape = predictor(img_frame, face)
        
        #검출된 랜드마크를 리스트로 저장
        list_points = []
        for p in shape.parts():
            list_points.append([p.x, p.y])
         
        #검출된 리스트를 numpy리스트로 변환   
        list_points= np.array(list_points)
        
        #검출된 랜드마크 중 index 변수에 지정된 부위만 이미지에 원으로 그려준다.
        for i,pt in enumerate(list_points[index]):
            
            pt_pos = (pt[0], pt[1])
            cv.circle(img_frame, pt_pos, 2, (255,255,255), -1)
         
        # 검출된 영역에서 붉은 사각형을 그리게 함
        cv.rectangle(img_frame, (face.left(), face.top()), (face.right(), face.bottom()), (0,0,255), 3)
        
    # 검출    
    cv.imshow('KooBonGeun', img_frame)
    
    # 키보드 입력을 받게함
    key = cv.waitKey(1)
    
    # esc키를 누르면 프로그램 종료
    if key == 27:
        break
        
    elif key == ord('1'):
        index = ALL
    elif key == ord('2'):
        index = LEFT_EYEBROW + RIGHT_EYEBROW
    elif key == ord('3'):
        index = LEFT_EYE + RIGHT_EYE
    elif key == ord('4'):
        index = NOSE
    elif key == ord('5'):
        index = MOUTH_OUTLINE + MOUTH_INNER
    elif key == ord('6'):
        index = JAWLINE
        
cap.release()
```
